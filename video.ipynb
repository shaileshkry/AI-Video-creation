```python
imgs=[]
for i in range(1,4):
    img='img'+str(i)+'.jpg'
    print(img)
    img=cv.imread(img)
    print(img.shape)
    imgs.append(img)
    #plt.imshow(img)
    
```

```python
plt.imshow(imgs[0])
```

```python
plt.imshow(imgs[1])
```

```python
imgs[1].shape
```

```python
resized_img=[]
i=1
for img in imgs:
    print(i)
    image=cv.resize(img,(400,400))
    resized_img.append(image)
    i+=1
```

```python
height,width,layers=resized_img[0].shape
print(height,width,layers)
```

```python
min=1
sec=54
video=cv.VideoWriter('advani.mp4',
                    cv.VideoWriter_fourcc(*'mp4v'),1,(width,height))
num_frames=(min*60+sec)*1
repeated_images = np.tile(resized_img, (num_frames // len(imgs), 1, 1,1 ))
for img in repeated_images:    
    video.write(img)
    #print(i)
    #print(img)
cv.destroyAllWindows()
video.release()
```
