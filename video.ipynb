```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
img='Img'+str(1)+'.jpg'
img=cv.imread(img)
```

```python
#print(img)
```

```python
imgs=[]
a=1
b=7
txt1="Gyanwapi"
txt2="Samachar Hut"
for i in range(a,b):
    if i>1:
        img2=img1
    else:
        img2='img'+str(1+(i)%(b-1))+'.jpg'
        #print(img2)
        img2=cv.imread(img2)
        img2=cv.resize(img2,(600,400))
    img1='img'+str(i)+'.jpg'
    #print(img)
    img1=cv.imread(img1)
    img1=cv.resize(img1,(600,400))
    
    
    #print(img1.shape)
    #imgs.append(img1)
    for k in range(1,40):
        if i%2==0:
            img=cv.addWeighted(img1,0.025*k,img2,1-0.025*k,0)
        else:
            img=cv.addWeighted(img1,1-0.025*k,img2,0.025*k,0)
        #adding text
        r=int((img1[350][350]+250)[0])
        g=int((img1[350][350]+250)[1])
        b=int((img1[350][350]+250)[2])
        cv.putText(img,txt1,(10*k,k*10),cv.FONT_HERSHEY_COMPLEX,1,
                   (r,g,b),1)
        cv.putText(img,txt2,(10*(40-k),(40-k)*10),cv.FONT_HERSHEY_COMPLEX,1,
                   (r,g,b),1)
        imgs.append(img)
    #plt.imshow(img)
    
```

```python
min=1
sec=15
fps=10
width=imgs[0].shape[1]
height=imgs[0].shape[0]
video=cv.VideoWriter('video.mp4',
                    cv.VideoWriter_fourcc(*'mp4v'),fps,(width,height))
num_frames=(min*60+sec)*fps
repeated_images = np.tile(imgs, (num_frames // len(imgs), 1, 1,1 ))
for img in repeated_images:    
    video.write(img)
    #print(i)
    #print(img)
cv.destroyAllWindows()
video.release()
```
