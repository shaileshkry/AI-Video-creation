```python
import requests as rq
from bs4 import BeautifulSoup as bsp
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize,sent_tokenize
import re
from nltk.corpus import stopwords
```

```python
#import nltk
#nltk.download('wordnet')
```

```python
stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()
```

```python
link='https://www.ndtv.com/india-news/lk-advani-to-be-conferred-bharat-ratna-announces-pm-modi-4985198'
```

```python
res=rq.get(link)
res
```

```python
html=res.text
```

```python
soup=bsp(html,'html.parser')
```

```python
soup
```

```python
txt=soup.find('p').text
txt=txt.replace('\n\n',' ').replace('\r',' ').replace('\xa0','')
txt
```

```python
txt=""""""
j=0
for i in soup.find_all('p'):
    if j<10:
        txt+=i.text
        print(i.text)
    j+=1
```

```python
txt
```

```python
stopwords.words()
```

```python
sentences=""""""

stem_sents=""""""
set_stem_sents=""""""

lemm_sents=""""""
set_lemm_sents=""""""

for sentence in sent_tokenize(txt):
    #print(sentence.lower())
    sentence=re.sub(r'[^\w\s]','',sentence)
    #print(sentence.split())
    words=""""""
    stem_words=""""""
    lemm_words=""""""
    set_stem_words=set()
    set_lemm_words=set()
    for word in  sentence.lower().split():
        if word not in stopwords.words() and len(word)>1:
            stem_words+=stemmer.stem(word) + ' '
            set_stem_words.add(stemmer.stem(word))
            
            lemm_words+=lemmatizer.lemmatize(word) + " "
            set_lemm_words.add(lemmatizer.lemmatize(word))
            
            words+=word+' '
    #print(words)
    
    sentences+= words+' '
    
    stem_sents+=stem_words + " "
    set_stem_sents+=" ".join(list(set_stem_words))
    
    lemm_sents+=lemm_words + " "
    set_lemm_sents+=" ".join(list(set_lemm_words))
    
print(sentences)
```
