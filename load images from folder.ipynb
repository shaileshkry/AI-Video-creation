```python
import matplotlib.pyplot as plt
import cv2 as cv
import os
```

```python
path=path='maps//train//'
```

```python
img_files=os.listdir(path)
len(img_files)
```

```python
images=[]
for image_file in img_files:
    img_path=os.path.join(path,image_file)
    img=plt.imread(img_path)
    img=cv.resize(img,(512,256))
    images.append(img)
```
